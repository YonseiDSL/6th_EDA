```python
# 댓글을 달 빈 리스트를 생성합니다.
Url_List = [] # URL 보관하는 리스트
Comments_count = 0


# 라이브러리를 로드합니다.
from bs4 import BeautifulSoup
import requests
import re
import sys
 
# 네이버 뉴스 url을 입력합니다.
Url_List.append("https://news.naver.com/main/read.naver?mode=LSD&mid=sec&sid1=102&oid=214&aid=0001138007")
Url_List.append("https://news.naver.com/main/read.naver?mode=LSD&mid=sec&sid1=104&oid=025&aid=0003121205")
Url_List.append("https://news.naver.com/main/read.naver?mode=LSD&mid=sec&sid1=105&oid=366&aid=0000749340")
Url_List.append("https://news.naver.com/main/read.naver?mode=LSD&mid=sec&sid1=104&oid=028&aid=0002553972")
Url_List.append("https://news.naver.com/main/read.naver?mode=LSD&mid=sec&sid1=104&oid=001&aid=0012548876")
Url_List.append("https://news.naver.com/main/read.naver?mode=LSD&mid=sec&sid1=104&oid=005&aid=0001459386")
Url_List.append("https://news.naver.com/main/read.naver?mode=LSD&mid=sec&sid1=104&oid=003&aid=0010626538")
Url_List.append("https://news.naver.com/main/read.naver?mode=LSD&mid=sec&sid1=101&oid=001&aid=0012548897")
Url_List.append("https://news.naver.com/main/read.naver?mode=LSD&mid=sec&sid1=102&oid=014&aid=0004680691")
Url_List.append("https://news.naver.com/main/read.naver?mode=LSD&mid=sec&sid1=102&oid=001&aid=0012547923")


for url in Url_List:
  List = []
  oid = url.split("oid=")[1].split("&")[0] #422
  aid = url.split("aid=")[1] #0000430957
  page = 1
  header = {
    "User-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.181 Safari/537.36",
    "referer": url,
  }
 
  while True:
      c_url = "https://apis.naver.com/commentBox/cbox/web_neo_list_jsonp.json?ticket=news&templateId=default_society&pool=cbox5&_callback=jQuery1707138182064460843_1523512042464&lang=ko&country=&objectId=news" + oid + "%2C" + aid + "&categoryId=&pageSize=20&indexSize=10&groupId=&listType=OBJECT&pageType=more&page=" + str(
          page) + "&refresh=false&sort=FAVORITE"
    # 파싱하는 단계입니다.
      r = requests.get(c_url, headers=header)
      cont = BeautifulSoup(r.content, "html.parser")
      total_comm = str(cont).split('comment":')[1].split(",")[0]
 
      match = re.findall('"contents":([^\*]*),"userIdNo"', str(cont))

    # 댓글을 리스트에 중첩합니다.
      List.append(match)

    # 한번에 댓글이 20개씩 보이기 때문에 한 페이지씩 몽땅 댓글을 긁어 옵니다.
      if int(total_comm) <= ((page) * 20):
        break
      else:
          page += 1
 
 
# 여러 리스트들을 하나로 묶어 주는 함수입니다.
  def flatten(l):
      flatList = []
      for elem in l:
          # if an element of a list is a list
          # iterate over this list and add elements to flatList
          if type(elem) == list:
              for e in elem:
                  flatList.append(e)
          else:
              flatList.append(elem)
      return flatList
 
 
# 리스트 결과입니다.
  allComments = flatten(List)
  Comments_count += len(allComments)
  sys.stdout = open('./DATA/7월 신문기사 댓글.txt','a')

print('총 누적 댓글갯수')
print(Comments_count, '개')
```

```python
# 필요한 라이브러리 import
!pip install konlpy
!pip install wordcloud
import konlpy
import re
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
from os import path
```

```python
# 월별 신문기사를 for문으로 차례차례 입력.
for i in range(2, 8):
  data = str(f"{i}월 신문기사 댓글.txt")
  with open(r'./DATA/' + data, encoding='cp949') as f:
    text = f.readlines()
  

  # <데이터 전처리>
  # 각 문서 마지막 두 문장 제외(필요없는 text)
  text = text[: -2]
  
  # |\n을 제거하는 것.
  text_strip = list([i.strip() for i in text if i != '\n'])
  text_join = ' '.join(text_strip)
  # 특수문자 제거
  filtered_content = re.sub(r'[^\d\s\w]', ' ', text_join)
  # 특수문자들을 공백으로 변환
  
  # Komoran을 사용해서 토큰화
  komoran = konlpy.tag.Komoran()
  komoran = konlpy.tag.Komoran(userdic='./DATA/NaverNewsDataDict.txt')
  komoran_nouns = komoran.nouns(filtered_content)

  # 불용어 제거
  stop_words = ['백신', '접종', '아스트라', '화이자', '얀센', '아스트라제네카', '모더나', '코로나', '나라', '우리나라', '기사', '댓글', '사람', '때문', '국민', '인간']
  for word in komoran_nouns:
    if word in stop_words or len(word) == 1:
        while word in komoran_nouns:
            komoran_nouns.remove(word)
  
  # <단어 빈도 분석>
  # Counter 사용해서 단어 빈도 분석
  c = Counter(komoran_nouns)

  print(f'<{i}월 신문기사 댓글 단어빈도 Top 20>')
  print(c.most_common(20))
  print()
  print(f'<{i}월 신문기사 댓글 워드 클라우드>')

  # <WordCloud 생성>
  img = plt.imread(r"./DATA/KoreaMap.jpg")
  wordcloud = WordCloud(
       font_path=FONT_PATH,
        background_color="white",
        random_state=1,
        colormap = 'Pastel1',
        color_func=ImageColorGenerator(img),      
        mask=img)
  cloud = wordcloud.generate_from_frequencies(c)
  
  # 생성된 WordCloud를 test.jpg로 보낸다.
  cloud.to_file('./DATA' + f'/{i}월 신문기사 댓글 워드 클라우드.jpg')
  
  #wordcloud.to_image()
  #plt.axis('off')
  #plt.imshow(cloud)
  #plt.show()
```
