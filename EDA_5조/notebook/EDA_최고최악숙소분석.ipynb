# 분석

## 전처리

```python
best = np.ones(airbnb_top.shape[0])
airbnb_top['group'] = best
```

```python
worst = np.zeros(airbnb_bottom.shape[0])
airbnb_bottom['group'] = worst
```

```python
airbnb2 = pd.concat([airbnb_top,airbnb_bottom])
```

범주형 자료로 변환

```python
airbnb2[['property_type','room_type','bed_type','cancellation_policy']] = airbnb2[['property_type','room_type','bed_type','cancellation_policy']].astype('category')
```

* Property_type : Apartment, House, Other 로

```python
airbnb2.loc[(airbnb2.property_type != 'Apartment') &(airbnb2.property_type != 'House'), 'property_type'] = 'Other'
```

```python
airbnb2.property_type.value_counts()
```

룸 정보들 새로운 변수로

```python
# X
```

"description" 변수를 길이로

```python
airbnb2['description'] = airbnb2['description'].astype(str)
```

```python
airbnb2['des_len'] = airbnb2.description.apply(lambda x: len(x))
```

## Best / Worst 숙소 특징

```python
#setting figure size for future visualizations
sns.set(rc={'figure.figsize':(10,8)})
sns.set_style('white')
```

**가격**

```python
# sns.violinplot(data=airbnb2, x="group", y="log_price")
sns.violinplot(data=airbnb2, x="group", y="log_price")
```

```python
grid = sns.FacetGrid(data=airbnb2, col='city')
grid.map(sns.violinplot, "group", "log_price")
```

Insight : 'Best 애들은 가격이 적당한 수준을 유지하나, Worst 애들은 너무 낮거나 너무 높은 경우 있음' $\rightarrow$ '가격을 적정수준에 맞추는 것이 필요'
* 가격의 분산이 높다 (객관화 부족)

**Host 관련**

```python
# sns.countplot(data=airbnb2, x="host_has_profile_pic", hue="group") # 다 True 임
sns.countplot(data=airbnb2, x="host_identity_verified", hue="group")
```

```python
sns.violinplot(data=airbnb2, x="group", y="host_response_rate")
```

```python
sns.violinplot(data=airbnb2, x="group", y="des_len")
```

Best/Worst Group 차이 _ 연속형 변수 중 host_response_rate 와 des_len 쓸 수 있겠음
